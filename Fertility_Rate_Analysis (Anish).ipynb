```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import seaborn as sns
```

*'Fertiliy Rate in poorly resourced countries tend to be higher, However, due to high maternal and prenatal mortality, There is a reduction in Birthrates'*

While this statement does not provide any statistical data, nor does it quanitfy 'poorly resourced countries', We can try and expand upon this statement and check the relation of Maternal and Prenatal Mortality with Birth Rate, and Fertility Rate for Poorly Resourced Countries.

For our given datasets, Countries have already been labelled as "High Income, Low Income, Upper Middle Income, Lower Middle Income"

We shall Compare the overall Fertility Rate of Poorly Resourced Countries i.e Lower Income Countries compared to the World average (Which should be higher.)

We shall also consider Maternal and Prenatal Mortality rates for these countries, which, according to the statement should be high while Birthrates should be decreasing.

```python
fr = pd.read_csv('Fertility_Rate.csv')
fr = fr.drop(columns=['Country Code','Indicator Name','Indicator Code'])
fr = fr.set_index('Country Name')
fr
```

```python
fr_flip = fr.transpose()
fr
```

#### Fertility Rate of Poor Countries Should be Higher

```python
figure(figsize=(15, 5), dpi=80)
plt.plot(fr_flip['Low income'],label = 'Fertility Rate of Low Income Countries')
plt.plot(fr_flip['World'],label = 'Fertility Rate World Average')
plt.xlabel('Years')
plt.ylabel('Fertility Rate')
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

### Similarly, the statement also states that maternal and infant mortality for such poor countries would be higher, causing a decrease in Birthrates

```python
mm = pd.read_csv('Maternal Mortality.csv')
mm = mm.drop(columns=['Country Code','Indicator Name','Indicator Code'])
mm = mm.set_index('Country Name')
mm_flip = mm.transpose()
mm_flip
```

```python
im = pd.read_csv('Infant Mortality.csv')
im = im.drop(columns=['Country Code','Indicator Name','Indicator Code'])
im = im.set_index('Country Name')
im_flip = im.transpose()
im_flip
```

### Infant Mortality for Poor Countries vs World

```python
figure(figsize=(15, 5), dpi=80)
plt.plot(im_flip['Low income'],label = 'Infant Mortality Rate of Low Income Countries')
plt.plot(im_flip['World'],label = 'Infant Mortality Rate World Average')
plt.xlabel('Years')
plt.ylabel('Infant Mortality Rate')
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

### Maternal Mortality Rate for Poor Countries vs World

```python
figure(figsize=(15, 5), dpi=80)
plt.plot(mm_flip['Low income'],label = 'Maternal Mortality Rate of Low Income Countries')
plt.plot(mm_flip['World'],label = 'Maternal Mortality Rate World Average')
plt.xlabel('Years')
plt.ylabel('Maternal Mortality Rate')
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
br = pd.read_csv('Birth Rate.csv')
br = br.drop(columns=['Country Code','Indicator Name','Indicator Code'])
br = br.set_index('Country Name')
br_flip = br.transpose()
br_flip
```

```python
figure(figsize=(15, 5), dpi=80)
plt.plot(br_flip['Low income'],label = 'Birth Rate of Low Income Countries')
plt.plot(br_flip['World'],label = 'Birth Rate World Average')
plt.xlabel('Years')
plt.ylabel('Birth Rate')
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
figure(figsize=(15, 4), dpi=80)
# plt.plot(mm_flip.loc['1990':'2020',['Low income']],label = 'Maternal Mortality Rate of Low Income Countries')
plt.plot(im_flip.loc['1990':'2020',['Low income']],label = 'Maternal Mortality Rate of Low Income Countries')
plt.plot(br_flip.loc['1990':'2020',['Low income']],label = 'Birth Rate of Low Income Countries')

plt.xlabel('Years')
plt.ylabel('Birth Rate')
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

```python
# figure(figsize=(15, 4), dpi=80)
# create figure and axis objects with subplots()
fig,ax = plt.subplots(figsize=(15, 4))
# make a plot
ax.plot(im_flip.loc['1990':'2020',['Low income']],label = 'Infant Mortality Rate of Low Income Countries', color = 'red')
# set x-axis label
ax.set_xlabel("Years",fontsize=14)
# set y-axis label
ax.set_ylabel("Infant Mortality Rate of low income",color="red",fontsize=14)


# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(br_flip.loc['1990':'2020',['Low income']],label = 'Birth Rate of Low Income Countries',color = 'blue')
ax2.set_ylabel("Birth Rate of Low Income",color="blue",fontsize=14)
plt.show()

#reference: https://cmdlinetips.com/2019/10/how-to-make-a-plot-with-two-different-y-axis-in-python-with-matplotlib/
```
