```python
# importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
# reading world fertility rate date
fr_rough_data = pd.read_csv("Global_FR_data/FR_data.csv",sep=",")
fr_rough_data
```

```python
# melting and reshaping the data
years = list(map(str,range(1960,2018)))
fr_data = pd.melt(fr_rough_data, id_vars=['Country Name','Country Code'], value_vars=years, var_name='year', value_name='FR', col_level=None)
fr_data
```

```python
# global trend for the fertility rate
global_fr_data = fr_data.groupby(['year']).mean()
global_fr_data.dtypes
plt.plot(global_fr_data)
```

```python
# trend of fertility rate in USA
usa_fr_data = fr_data[fr_data['Country Code'] == 'USA']
plt.plot(usa_fr_data['year'], usa_fr_data['FR'])
```

Hypothesis 1: In larger metropolitan areas, the higher the immigrant share of the population, the fewer children native-born women have. This is the case even after controlling for age, education, income, race, marital status, and other factors.
There are a possible reasons why immigration might lower the fertility of native-born Americans -
It drives up housing costs for families with children, 
Lowers wages for some American workers, or strains public services relied on by families, such as local schools,  All of these factors may make American couples more reluctant to have children.

### Null Hypothesis: Immigration reduces tge native total fertility rate in US due increase in the home ownership of immigrants and increase in the use of public services by immigrants

### Alternate Hypothesis: Total Fertility Rate of the natives in US is independent of the immigration.
